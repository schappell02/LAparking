```python
pc_class = sclapc.LAparkC('sc_pc_2015_2018.csv')
```

Class of object that deals with parking citation data

```python
pc_class.plotD()
```

Plots the lognormal density of parking citations

```python
pc_class.plotD(PoI='UCLA',uagent=uagent)
```

Can plot point of interest, plotD function will call getLaLo if string is inputed as point of interest.

```python
pc_class.plotD([-118.267052602794, 34.0430046],uagent=uagent)
```

Can also give it list of longitude and latitud (in that order)

```python
pc_class.plotD(DoW='Monday')
```

```python
pc_class.plotD(DoW='Weekday')
```

Can also search by specific day (all 7 days of the week are an option), or by Weekday or Weekend.

```python
pc_class.plotD(DoW='Weekday',ToD='Early')
```

And can add a time of day requirement (Early, Morning, Afternoon, and Evening)

```python
pc_class.pcr_poi('Staples Center',uagent=uagent,dyrs=4)
```

```python
pc_class.pcr_poi('UCLA',DoW='Weeday',ToD='Morning',fmil=0.5,uagent=uagent,dyrs=4)
```

pcr_poi will give parking citation rate (number of parking citations per day for every square mile) at a given point of interest. Also returns highest parking citation rate for comparison.

Day of Week and Time of Day function as before. dyrs is the number of years over which the given dataset spans (4 in this case, set to 1yr unless otherwise states).

And can also change the precision of bins used as a fraction of a mile (fmil, set to 1/4 of a mile unless otherwise stated).
