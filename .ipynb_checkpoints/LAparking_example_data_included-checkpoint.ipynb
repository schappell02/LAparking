```python
uagent = #!!!!! Input own user agent, make it unique and specific
sclapc.getLaLo('UCLA',uagent=uagent)
```

getLaLo returns longitude and latitude for input search (uses geopy geocoders), can handle many points of interest and street address searches. However, it cannot handle intersection searches.

Note: getLaLo also adds 'CA' to the string search, so no need to add the state to the search.

```python
reload(sclapc)
pc_class = sclapc.LAparkC('sc_pc_2018only.csv')
```

```python
pc_class.plotD(nbins=300)
```

```python
pc_class.plotD(PoI=[-118.267052602794, 34.0430046],uagent=uagent,DoW='Weekend',ToD='Evening',nbins=100)
```

Can change number of bins with nbins

DoW can be used to select specific day of week (Monday .. Sunday) or Weekday or Weekend. DoT can specify which portion of the day to plot (Early, Morning, Afternoon, and Evening).

Point of Interest (PoI) can be list of longitude and latitude, or str (which will then be searched using geopy).

```python
pc_class.plotD(PoI='Venice Blvd., Los Angeles',DoW='Weekday',uagent=uagent,nbins=200)
```

```python
pc_class.pcr_poi('Staples Center',DoW='Weekend',ToD='Evening',uagent=uagent)
```

pcr_poi returns the parking rate (number of citations per day per square mile) for a given point of interest (can be list of longitude and latitude or a string search). DoW and ToD function as before.

```python
pc_class.pcr_poi('Staples Center',DoW='Weekday',uagent=uagent,fmil=0.5)
```

Can also change the precision, fmil (fraction of a mile).
