```python
import sc_LApc as sclapc
```

This notebook shows plots using my sc_LApc code and the parking citation information (provided by the city of Los Angeles, CA) between 2015 and 2018. Resulting csv file is too large to upload to Github

```python
pc_class = sclapc.LAparkC('sc_pc_2015_2018.csv')
```

Class of object that deals with parking citation data

```python
pc_class.plotD()
```

Density map of parking tickets between 2015 and 2018

```python
pc_class.plotD(PoI='7080 Hollywood Blvd., Los Angeles')
```

```python
pc_class.plotD(PoI='Runyon Canyon Park')
```

Can also handle coordinates, input as: [longitude, latitude]

```python
pc_class.plotD(DoW='Monday')
```

```python
pc_class.plotD(DoW='Weekday')
```

Can also search by specific day (all 7 days of the week are an option), or by Weekday or Weekend.

```python
pc_class.plotD(DoW='Weekday',ToD='Early')
```

And can add a time of day requirement (Early, Morning, Afternoon, and Evening)

```python
pc_class.pcr_poi('7080 Hollywood Blvd., Los Angeles')
```

```python
pc_class.pcr_poi('National Blvd., Los Angeles')
```

```python
pc_class.pcr_poi('7080 Hollywood Blvd., Los Angeles',DoW='Friday',ToD='Afternoon')
```

```python
pc_class.pcr_poi('National Blvd., Los Angeles',DoW='Friday',ToD='Afternoon')
```

Also function to return coordinates for street address or point of interest

```python
sclapc.getLaLo('7080 Hollywood Blvd., Los Angeles')
```
